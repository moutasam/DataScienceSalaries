### 3. Tuning and Cleaning

#### 3.1 Job Titles

```python
# Getting the total count per job title 

print(df.job_title.value_counts())


# Grouping

# Data Scientists

df.loc[df['job_title'].str.contains('Data Scientist'), 'job_title'] = 'Data Scientist'
df.loc[df['job_title'].str.contains('Research Scientist'), 'job_title'] = 'Data Scientist'
df.loc[df['job_title'].str.contains('Data Science Consultant'), 'job_title'] = 'Data Scientist'
df.loc[df['job_title'].str.contains('AI Scientist'), 'job_title'] = 'Data Scientist'

# Manager

df.loc[df['job_title'].str.contains('Manager'), 'job_title'] = 'Manager'
df.loc[df['job_title'].str.contains('Head'), 'job_title'] = 'Manager'
df.loc[df['job_title'].str.contains('Director'), 'job_title'] = 'Manager'
df.loc[df['job_title'].str.contains('Lead'), 'job_title'] = 'Manager'

# Data Analyst

df.loc[df['job_title'].str.contains('Data Analyst'), 'job_title'] = 'Data Analyst'
df.loc[df['job_title'].str.contains('Analytics Engineer'), 'job_title'] = 'Data Analyst'

# Data Engineer

df.loc[df['job_title'].str.contains('Data Engineer'), 'job_title'] = 'Data Engineer'
df.loc[df['job_title'].str.contains('Data Architect'), 'job_title'] = 'Data Engineer'
df.loc[df['job_title'].str.contains('Data Science Engineer'), 'job_title'] = 'Data Engineer'
df.loc[df['job_title'].str.contains('Data Analytics Engineer'), 'job_title'] = 'Data Engineer'
df.loc[df['job_title'].str.contains('Big Data Architect'), 'job_title'] = 'Data Engineer'



# Machine Learning Engineer

df.loc[df['job_title'].str.contains('Machine Learning'), 'job_title'] = 'Machine Learning Engineer'
df.loc[df['job_title'].str.contains('ML Engineer'), 'job_title'] = 'Machine Learning Engineer'
df.loc[df['job_title'].str.contains('Computer Vision Engineer'), 'job_title'] = 'Machine Learning Engineer'
df.loc[df['job_title'].str.contains('NLP Engineer'), 'job_title'] = 'Machine Learning Engineer'


# Machine Learning Engineer

df.loc[df['job_title'].str.contains('Machine Learning'), 'job_title'] = 'Machine Learning Engineer'
df.loc[df['job_title'].str.contains('ML Engineer'), 'job_title'] = 'Machine Learning Engineer'
df.loc[df['job_title'].str.contains('Computer Vision Engineer'), 'job_title'] = 'Machine Learning Engineer'
df.loc[df['job_title'].str.contains('NLP Engineer'), 'job_title'] = 'Machine Learning Engineer'

# others

df.loc[df['job_title'].str.contains('Computer Vision Software Engineer'), 'job_title'] = 'Others'
df.loc[df['job_title'].str.contains('Data Specialist'), 'job_title'] = 'Others'
df.loc[df['job_title'].str.contains('ETL Developer'), 'job_title'] = 'Others'
df.loc[df['job_title'].str.contains('3D Computer Vision Researcher'), 'job_title'] = 'Others'




print("\n \n \n")
print("The new job titles groups are:" , "\n")
print(df.job_title.value_counts())

```

#### 3.1 Experience

```python

print(df.experience_level.value_counts())


# Renaming letters to words

df.loc[df['experience_level'].str.contains('EN'), 'experience_level'] = 'Entry'
df.loc[df['experience_level'].str.contains('MI'), 'experience_level'] = 'Mid'
df.loc[df['experience_level'].str.contains('SE'), 'experience_level'] = 'Senior'
df.loc[df['experience_level'].str.contains('EX'), 'experience_level'] = 'Executive'


print("\n \n")
print("The new Experinece titles groups are:" , "\n")
print(df.experience_level.value_counts())
```
